 **DISCOUNT PERCENTAGE**

 Lets calculate the discount percentage on the items.

```python

# Calculate discount percentage
def calculate_discount_percentage(x):
    if x['LISTPRICE'] == 0:
        return 0
    return 100*(1-x['PRICE']/x['LISTPRICE'])
filtered_data['discount_percentage'] = filtered_data[['PRICE', 'LISTPRICE']].apply(calculate_discount_percentage, axis=1)


# Select columns
selected_cols = ['discount_percentage', 'TITLE', 'STARS', 'REVIEWS', 'PRICE', 'ISBESTSELLER', 'BOUGHTINLASTMONTH','CATEGORY_ID','CATEGORY_NAME']
filtered_data = filtered_data[selected_cols]

print(filtered_data.shape)
```

```python
# Calcular la matriz de correlación para el subconjunto de datos
correlation_matrix = filtered_data.corr()
print(correlation_matrix)
```

```python

# Crear un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', linewidths=0.5)
plt.title('CORRELATION MATRIX')
plt.show()
```

**Discount Percentage and STARS:** There is a very weak and positive correlation (0.031) between the discount percentage and product ratings (STARS). This suggests that, overall, the discount percentage has a minimal influence on product ratings.

**STARS and REVIEWS:** The correlation between product ratings (STARS) and the number of reviews (REVIEWS) is extremely low (0.004). This indicates that product ratings are not strongly related to the quantity of reviews they receive.

**PRICE and Discount Percentage:** There is a weak negative correlation (-0.048) between the price (PRICE) and the discount percentage. This suggests that, in general, products with higher discounts tend to have slightly lower prices.


**STARS and BOUGHTINLASTMONTH:** There is a positive correlation (0.084) between product ratings (STARS) and sales in the last month (BOUGHTINLASTMONTH). This suggests that, overall, products with higher ratings tend to sell better in the last month.

**DISCOUNT_PERCENTAGE and BOUGHTINLASTMONTH:** There is also a positive correlation (0.075) between the discount percentage (DISCOUNT_PERCENTAGE) and sales in the last month. This indicates that products with higher discounts tend to have higher sales in the last month.

**ISBESTSELLER and BOUGHTINLASTMONTH:** The variable ISBESTSELLER has a moderately positive correlation (0.235) with sales in the last month. This means that products considered best-sellers tend to have significantly higher sales in the last month.

**PRICE and BOUGHTINLASTMONTH:** The relationship between price (PRICE) and sales in the last month is negative, although the correlation is low (-0.014). This suggests that products with higher prices may have slightly lower sales in the last month, but the relationship is not strong.

Overall, most of the correlations in this matrix are very weak, indicating that the variables are not strongly related to each other. This means that product ratings are not strongly influenced by the discount percentage, reviews, or price. Additionally, best-selling products tend to be related to recent sales, but the relationship is not extremely strong.

Focusing on the relationship between sales (BOUGHTINLASTMONTH) and the rest of the variables in the dataset, we can observe the following:
In summary, product ratings (STARS), the discount percentage, whether a product is a best-seller, and price appear to be related in some way to sales in the last month. Products with higher ratings, higher discounts, and those considered best-sellers tend to have higher sales in the last month. However, the relationship with price is less clear and does not strongly influence sales. These observations can be helpful in understanding the factors that influence product sales in your dataset.
